**Project Title:** Dividend Yield Fund  
**Group :** 31  

## Overview  
This project analyzes the performance of a Dividend Yield Fund using historical stock and dividend data. It focuses on computing dividend yields, evaluating portfolio returns, and visualizing analytical insights using Python and Jupyter Notebook.

## How to run  
1. Create a virtual environment:  
   `python -m venv venv`

2. Activate the environment:  
   - Windows: `venv\Scripts\activate`  
   - macOS/Linux: `source venv/bin/activate`

3. Install dependencies:  
   `pip install -r requirements.txt`

4. Open and run the notebook:  
   `jupyter notebook`  
   Then open **Group31_Code.ipynb**
